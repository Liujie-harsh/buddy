```python
# 贝叶斯算法示例：朴素贝叶斯分类器用于文本分类

from sklearn.feature_extraction.text import CountVectorizer# 文本特征提取
from sklearn.naive_bayes import MultinomialNB# 朴素贝叶斯分类器

# 示例数据
texts = [
    "我喜欢看电影",
    "这部电影很好看",
    "我不喜欢这部电影",
    "这部电影很无聊",
    "我觉得这部电影很精彩",
    "这部电影太差了"
]
labels = [1, 1, 0, 0, 1, 0]  # 1表示喜欢，0表示不喜欢

# 文本特征提取
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
print(X.toarray())

# 训练朴素贝叶斯分类器
clf = MultinomialNB()
clf.fit(X, labels)

# 测试新样本
test_texts = ["这部电影非常精彩", "这部电影很无聊"]
X_test = vectorizer.transform(test_texts)
pred = clf.predict(X_test)

for text, label in zip(test_texts, pred):
    print(f"'{text}' => {'喜欢' if label == 1 else '不喜欢'}")
```
